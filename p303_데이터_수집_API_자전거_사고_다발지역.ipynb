```python
result = []
data = response.json()
w_data = data['items']['item']
for row in range(len(w_data)):
  line={}  #딕셔너리 생성
  line['spot_nm'] = w_data[row]['spot_nm']  #딕셔너리 값 넣어주기
  line['lo_crd'] = w_data[row]['lo_crd']
  line['la_crd'] = w_data[row]['la_crd']
  result.append(line)
```

```python
result
```

```python
#저장
# indent="\t" : 들여쓰기를 탭으로 설정
# ensure_ascii=False : 유니코드로 변환되는 작업 방지

with open('./cycle.json', 'w', newline='', encoding='utf-8') as f:
  json.dump(result, f, indent="\t", ensure_ascii=False)
```

```python
#딕셔너리 형태의 문자열'''
json_str='''
  {
    "name":"홍길동",
    "age":"30"
  }'''
```

```python
type(json_str)
```

json.load : 파일 내용 읽어오기  
json.dump : 파일에 저장하기  
json.loads : 문자열(str) -> 객체(list, dict, tuple...)  
json.dumps : 객체(list, dict, tuple...) -> 문자열(str)  



```python
import json

#str -> dict
data = json.loads(json_str)   #loads에는 딕셔너리 형태의 문자열을 넣어줌
data  #딕셔너리로 바꿔서 리턴
```

```python
type(data)
```

```python
# dict -> str
sn = json.dumps(data, ensure_ascii=False)  #ensure_ascii=False:유니코드로 변경 방지
sn  #문자열로 리턴
```

```python
type(sn)
```

```python
json_str2 = '[1,2,3,4]'

# str -> list
data = json.loads(json_str2)
print(data)
print(type(data)) #리스트로 리턴(객체로 바뀜)
```

```python
# list -> str
sn = json.dumps(data)
print(sn)
print(type(sn))  #객체를 문자열로
```

loads , dumps 등 뒤에 s가 붙어있는 것은 string의 뜻. 문자열과 관련
