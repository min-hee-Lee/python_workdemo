### pivot
- 각 컬럼의 값을 교차테이블 구성요소로 전달, 교차테이블 완성
- index, columns, values 컬럼을 각각 전달
- 요약기능 없음
- dataframe.pivot(index,columns,values)

```python
krecom_pivot = krecom_data.pivot(index='UserID', columns='MovieID', values='Rating')
krecom_pivot
```

```python
# inplace=True는 실제 데이터에 변경사항이 적용된다.

#krecom_pivot.fillna(method='ffill') #method='ffill' or method='pad'  # 결측값을 앞에 있는걸로 채워라
#krecom_pivot.fillna(method='bfill')  #method='bfill' or method=backfill' #결측값을 뒤에 있는 걸로 채워라
#krecom_pivot.fillna(krecom_pivot.mean(), axis=1)  #결측값을 평균을 내서 채워라
krecom_pivot.fillna(0, inplace=True)  #결측값을 0으로 채워라(null값을 0으로 채워줘라)
```

### pivot_table
- pivot 기능과 유사, 더 많은 옵션 사용 가능
- index,columns,values 컬럼을 각각 전달
- aggfunc 옵션을 사용해 요약 기능 전달 가능(기본은 평균)
- full_value옵션 사용해 NA값 대체 가능
- dataframe.pivot_table(index, columns, values, aggfunc='mean', fill_value)

```python
#최소값 구하기
krecom_pivot_table = krecom_data.pivot_table(index='UserID', columns='MovieID', values='Rating', aggfunc='min', fill_value=0)
krecom_pivot_table
```

```python
df = pd.DataFrame(data= {'A':['foo', 'foo', 'foo', 'foo', 'foo', 'bar', 'bar', 'bar', 'bar'],
                         'B':['one', 'one', 'one', 'two', 'two', 'one', 'one', 'two', 'tow'],
                         'C':['small', 'large', 'large', 'small', 'small', 'large', 'small', 'small', 'large'],
                         'D':[1, 2, 2, 3, 3, 4, 5, 6, 7],
                         'E':[2, 4, 5, 5, 6, 6, 8, 9, 9]})
print(df)
```

```python
#그룹별 평균 구하기
table = pd.pivot_table(df, values=['D','E'], index=['A','C'])
table
```

```python
table = pd.pivot_table(df, values=['D','E'], index=['A', 'C'], aggfunc={'D' : np.mean,              #D의 평균값
                                                                        'E':[min, max, np.mean]})   #E의 최소값, 최대값, 평균값
table
```

### stack
- unstack은 index를 column으로 만들어준다.
- stack은 column을 index로 만들어 준다.
- 이미지 참고 : https://www.w3resource.com/pandas/dataframe/dataframe-stack.php


```python
df_single_level_no_index = pd.DataFrame([[0,2],[3,4]],
                                        columns=['weight', 'height'])
df_single_level_no_index
```

```python
df_single_level_no_index.stack()
```

```python
df_single_level_no_index.stack().unstack() 
```

```python
df_single_level_cols = pd.DataFrame([[0,2],[3,4]],
                                    index=['deer','monkey'],
                                    columns=['weight', 'height'])
df_single_level_cols
```

```python
df_single_level_cols.stack()
```

```python
df_single_level_cols.stack().unstack()
```

```python
multicol1 = pd.MultiIndex.from_tuples([('weight', 'kg'),
                                       ('weight', 'pounds')])
df_multi_level_cols1 = pd.DataFrame([[3, 4], [4, 5]],
                                    index=['deer', 'monkey'],
                                    columns=multicol1)
df_multi_level_cols1
```

```python
df_multi_level_cols1.stack()
```

```python
#결과는 위와 같음
df_multi_level_cols1.stack(1)
```

```python
df_multi_level_cols1.stack(0)
```

```python
df_multi_level_cols1.stack(0).unstack()
```

```python
# swaplevel 열이나 행 레벨 바꾸기 (axis=1 열(컬럼) / axis=0 행(인덱스))
swap = df_multi_level_cols1.stack(0).unstack()
swap.swaplevel(0,1, axis=1) #column의 계층(level) 순서 변경
```
